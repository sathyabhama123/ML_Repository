**Missing values**


```python
data['criminal_cases'].value_counts()
```

```python
data.isna().sum()
```

```python
data['criminal_cases'] = data['criminal_cases'].replace({'Not Available': 0})
data['criminal_cases'] = data['criminal_cases'].fillna(0)
```

```python
data.isna().sum()
```

```python
plt.hist(data['age'])
plt.show()
```

```python
data["age"]=data["age"].fillna(data["age"].mean())
```

```python
#Replace multiple words in a column for certain words.
def update_education(row):
    if row['education'] in ['Illiterate','Not Available']:
        return 'Iliterate/NA'
    if row['education'] in ['5th Pass','8th Pass','10th Pass','12th Pass']:
        return 'School Level'
    elif row['education'] in ['Graduate Professional','Graduate']: 
        return 'Graduate'
    elif row['education'] in ['Post Graduate\n','Post Graduate', 'Doctorate']:
        return 'Post Graduate/Doctorate'
    else:
        'Literate'
        
data['education'] = data.apply(update_education,axis =1)
```

```python
data.education.unique()
```

```python
#Replace NA in education wth Iliterate/NA.
data['education'].fillna('Iliterate/NA', inplace=True)
```

```python
data.isna().sum()
```

```python
for i in ['symbol','gender','category','assets','liabilities']:
  data[i]=data[i].fillna(data[i].mode)
```

**Logistic Regression**

```python
from sklearn.linear_model import LogisticRegression
logit_model=LogisticRegression()
logit_model.fit(X_train,y_train)
y_pred=logit_model.predict(X_test)
y_pred
```

**KNN**

```python
from sklearn.neighbors import KNeighborsClassifier
metric_k=[]
neighbors=np.arange(3,10)
```

```python
for k in neighbors:
    classifier = KNeighborsClassifier(n_neighbors=k,metric='minkowski',p=2)
    classifier.fit(X_train,y_train)
    y_pred=classifier.predict(X_test)
    acc=accuracy_score(y_test,y_pred)
    metric_k.append(acc)
```

```python
plt.plot(neighbors,metric_k,'o-') 
plt.xlabel('k value')
plt.ylabel('accuracy')
plt.grid()
```

```python
for k in neighbors:
    classifier = KNeighborsClassifier(n_neighbors=9,metric='minkowski',p=2)
    classifier.fit(X_train,y_train)
    y_pred=classifier.predict(X_test)
    acc=accuracy_score(y_test,y_pred)
    metric_k.append(acc)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score
c_matrix=confusion_matrix(y_test,y_pred)
print('Confusion matrix of kNN classification  is',c_matrix)
```
